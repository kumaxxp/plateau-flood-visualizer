# 推移閾値で水没面を作る

「任意の水位（例：T.P.+X m）で水面を上げ、地形（DEM）と比較して“水没域”を算出→重ね描画」というやり方。こちらはDEMレイヤが必要です。

```python
!plateaukit prebuild plateau-13101-chiyoda-ku-2023 -t dem -t tran
```

エリアを切る → DEM から標高グリッドを得て、標高 < 水位 をマスク → ポリゴン化。
そのポリゴンを水面レイヤとして area.show() に重ねる（半透明の青など）。

```python
area = ds.area_from_landmark("東京駅")
gdfs = area.to_geodataframe()
dem = gdfs["dem"]        # DEM（格子 or TIN → PlateauKitがGeoDataFrame/GeoSeriesで返す形に依存）

WATER_LEVEL = 5.0        # 例: T.P.+5 m とする（標高系はデータの定義に合わせること）

# ① DEM をサンプル → ② 標高 < 水位 の領域をラスタ→ベクタ化（iso面）
#    ここは環境に応じて：geopandas+shapely か、raster化→等値面抽出を使う

water_polygons = compute_inundation_polygons_from_dem(dem, WATER_LEVEL)  # ユーザー関数

# ③ 表示（建物＆道路＋“水没域”）
area.show(
    layers=["bldg","tran"],
    overlays=[{"name":"water","geometry": water_polygons, "style":{"fillOpacity":0.4}}]
)
```
